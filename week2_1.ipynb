# week 2: Lotka-Vorterra equation

    - dP/dt = aP - bPZ
    - dZ/dt = bPZ - cZ

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tendency(t,Xin,a,b,c):
    # unpack input
    p = Xin[0]
    z = Xin[1]
    # calculate tendency
    dp = a*p-b*p*z
    dz = b*p*z-c*z
    # output tendency
    Xout=np.array([dp,dz])
    return Xout
```

```python
# set parameters
a=1
b=1
c=1
dt=.1
po=.5
zo=.5
```

```python
# set time
t=np.arange(0,50+dt,dt)
N=np.size(t)
```

```python
# time stepping loop
P=np.zeros(N)
Z=np.zeros(N)
P[0]=po
Z[0]=zo
for n in range(N-1):
    Xin=np.array([P[n],Z[n]])
    Xout=tendency(t[n],Xin,a,b,c)
    P[n+1]=P[n]+dt*Xout[0]
    Z[n+1]=Z[n]+dt*Xout[1]
```

```python
plt.plot(t,P,label='P')
plt.plot(t,Z,label='Z')
```
