# observational datasets for ecosystem modeling
    - Satellite NPP

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
# Satellite NPP
ds=xr.open_dataset('NPP_VGPM_MODISR18_3x3_2003_2018.nc')
```

```python
ds
```

```python
ds.time
```

```python
ds.NPP.isel(time=0).plot(vmin=0,vmax=1000)
```

```python
ds.NPP.sel(lat=40,method='nearest').sel(lon=-130,method='nearest').plot()
```

```python
ds.NPP.sel(lat=40,method='nearest').sel(lon=-130,method='nearest').plot(label='130W')
ds.NPP.sel(lat=40,method='nearest').sel(lon=-160,method='nearest').plot(label='160W')
plt.legend()
plt.show()
```

```python
# calculate climatology
NPPC = ds.NPP.groupby('time.month').mean('time')
```

```python
NPPC
```

```python
NPPC.sel(month=4).plot(vmax=1000)
```

```python
NPPC.sel(lat=40,method='nearest').sel(lon=-130,method='nearest').plot()
```

```python
# create a numpy array
NPPC_array = NPPC.to_numpy()
```

```python
NPPC_array.shape
```

```python
# save a numpy array
np.save('NPPC.npy',NPPC_array)
nppc=np.load('NPPC.npy')
```

```python
# save a xarray file, first create a xarray "dataset"
dsc = NPPC.to_dataset()
```

```python
dsc
```

```python
# writing a netCDF file is easy
dsc.to_netcdf('NPP_VGPM_3x3_climatology.nc')
```
